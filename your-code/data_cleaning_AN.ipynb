#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import the users table.

```python
users_path = r"C:\Users\Working\Desktop\Ironhack_projects\lab-data-cleaning\data\users.csv"

df_users = pd.read_csv(users_path)

df1_users = df_users.copy()
df1_users
```

#### 3. Rename Id column to userId.

```python
df1_users.rename(columns={"Id":"userId"}, inplace=True)
df1_users
```

#### 4. Import the posts table. 

```python
posts_path = r'C:\Users\Working\Desktop\Ironhack_projects\lab-data-cleaning\data\posts.csv'

df_posts = pd.read_csv(posts_path)
df1_posts = df_posts.copy()

df1_posts
```

#### 5. Rename Id column to postId and OwnerUserId to userId.

```python
df1_posts.rename(columns={'Id':'postId', 'OwnerUserId' : "userId"}, inplace=True)
df1_posts
```

#### 6. Define new dataframes for users and posts with the following selected columns:
**users_sliced columns**: userId, Reputation, Views, UpVotes      
**posts_sliced columns**: postId, Score, userId, ViewCount

```python
user_sliced = df1_users.loc[:, ['userId', 'Reputation', 'Views', 'UpVotes']]
user_sliced.head()
```

```python
posts_sliced = df1_posts.loc[:, ['postId', 'Score', 'userId', 'ViewCount']]
posts_sliced.head()
```

```python
print(f'user_sliced shape: {user_sliced.shape}')
print(f'posts_sliced shape: {posts_sliced.shape}')
```

#### 7. Merge the two dataframes created in the step above (8), users_sliced and posts_sliced. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df_user_sliced_and_posts_sliced_merged = pd.merge(user_sliced, posts_sliced)
df_user_sliced_and_posts_sliced_merged.head()
```

Basically there are three options:

1. Refilling - If we try to refill the variable, that means doing a many correlation analysis with other variables and hoping that it will correlate highly wit any of them, so that we can do a regression analysis and add the new values. Considering that we do not have many other variables, we have low chances to find a good enough correlation in order to explain the addition of our new values. 

2. Keep the variable - If we keep the values, we can do a half-analysis, but that means removing more than 50% of values from all other variables. Tottaly not recomended

3. Save the variable and remove it - We save the variable in a separate container in case someday we manage to recover the missing information, then we remove the variable and excluded from current analysis.

I will stick to the options **#3**, because it seem to me to be the most logical in this case.
